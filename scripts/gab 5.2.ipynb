```python
#Chargement du geopandas
f_vege=gpd.read_file('data\project\FORMATION_VEGETALE.shp')
f_vege.head()
```

```python
#Création de la liste pour le masque
L_mask=['Lande','Formation herbacée','Forêt ouverte de conifères purs','Forêt ouverte de feuillus purs','Forêt ouverte sans couvert arboré',
        'Forêt ouverte à mélange de feuillus et conifères','Forêt fermée sans couvert arboré']
```

Ci dessous il y a initialisation d'un numpy de un de même taille que le tableau pour le concatener à notre tableau

```python
ones=np.ones((24041,1),dtype=int)
f_vege['mask']=ones
f_vege.head()
```

Ici on construit la colonne qui vaut 1 si on est de la forêt et 0 sinon

```python
for i,j in zip(f_vege['TFV'],range(len(f_vege['mask']))):
    if i in L_mask:
        #f_vege['mask'][j]=0
        f_vege.loc[j,'mask']=0
f_vege.head()
```

Ici c'est l'étape pour avoir les même valeurs que le prof

```python
f_vege['Valeur du pixel']=f_vege['mask']
f_vege.head()
```

Ici je crée la deuxième colonne correspondant à classe (étapes faisable direct avant dans la création du mask)
Il y a aussi création du nouveau tableau avec juste le masque et l'ID des polygons pour plus tard

```python
for i in range(len(f_vege['mask'])):
    if f_vege['Valeur du pixel'][i]==1:
        f_vege.loc[i,'mask']='Zone de forêt'
        #f_vege['mask'][i]='Zone de forêt'
    else:
        f_vege.loc[i,'mask']='Zone hors forêt'
        #f_vege['mask'][i]='Zone hors forêt'
f_vege['Classe']=f_vege['mask']
Masque=f_vege[['ID','Classe','Valeur du pixel']]
Masque.head()
```

```python
#Encodage en 8 bits
Masque=Masque['Valeur du pixel'].astype('uint8')
```
